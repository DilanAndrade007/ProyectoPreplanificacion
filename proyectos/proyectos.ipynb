Ahora vamos a cambiar la fecha de inicio del proyecto por la fecha de inicio del periodo académico para poder hacer el cálculo de semanas y horas por semana

Queremos tomar las horas aprobadas y dividir para 24, que son las semanas que hay dentro del periodo, entonces obtenemos las horas que debe hacer por semana 

```python
df["HORAS SEMANALES"] = df["HORAS APROBADAS"] / 24

df.head()
```

Vamos a definir las fechas del periodo académico para poder tener claro el periodo de tiempo en que se aplican las horas aprobadas

```python
# 2) Definir el periodo académico
inicio_periodo = pd.Timestamp("2025-09-01")
fin_periodo    = pd.Timestamp("2026-02-28")
```

Ahora con las fechas del periodo establecidas, podemos poner las restricciones a las fechas de los proyectos, por ejemplo, verificar que el proyecto esté dentro de este periodo y no se encuentre en fechas fuera de este

```python
# Asegurar tipos datetime
df["Fecha inicio"] = pd.to_datetime(df["Fecha inicio"], errors="coerce")
df["Fecha fin real"] = pd.to_datetime(df["Fecha fin real"], errors="coerce")

# Regla solicitada:
# 1) Si inicio < inicio_periodo y fin < inicio_periodo  => "Fuera del periodo académico"
fuera_periodo = (df["Fecha inicio"] < inicio_periodo) & (df["Fecha fin real"] < inicio_periodo)
```

```python
# Crear/actualizar observaciones
df["observaciones"] = ""
df.loc[fuera_periodo, "observaciones"] = "Fuera del periodo académico"

# (Opcional) Si quieres ver un conteo:
print("Marcados como fuera del periodo académico:", fuera_periodo.sum())

df.head()
```

Es tiempo de hacer los cálculos de los días disponibles en el periodo del proyecto. Lo primero será recortar las fechas a las del periodo. Si el proyecto empieza antes, entonces se considera desde el comienzo del periodo. Si el proyecto termina despues, se considera el fin del periodo en lugar del fin del proyecto.

```python
# 4) Recorte de fechas a los límites del periodo
inicio_clip = df["Fecha inicio"].where(df["Fecha inicio"] >= inicio_periodo, inicio_periodo)
fin_clip    = df["Fecha fin real"].where(df["Fecha fin real"] <= fin_periodo,    fin_periodo)
```

Ahora sí, es momento de hacer el cálculo respectivo ya con las fechas solapadas.

```python
# 5) Calcular días solo donde hay solape y fechas válidas
fechas_validas = df["Fecha inicio"].notna() & df["Fecha fin real"].notna()
solapa = (fin_clip >= inicio_clip)
```

```python
# Días inclusivos: fin - inicio + 1
dias = (fin_clip - inicio_clip).dt.days + 1
dias = dias.where(solapa, 0)             # si no hay solape, 0 días
dias = dias.where(fechas_validas)        # si faltan fechas, NaN
dias = dias.where(~fuera_periodo)        # si está fuera del periodo académico, NaN

# 6) Asignar columna (entero con soporte de NaN)
df["Dias disponibles"] = dias.astype("Int64")

# Vista rápida
df[["Fecha inicio", "Fecha fin real", "Dias disponibles", "observaciones"]].head()
```

Con los días calculados vamos a ver cuántas semanas tenemos entonces, recordamos que se debe dividir para 5 y no para 7

```python
# 1) Semanas disponibles (pueden ser fraccionarias)
df["semanas_disponibles"] = (df["Dias disponibles"] / 5).astype(float)

# 2) Horas disponibles en el periodo (semanas * horas semanales)
df["horas_disponibles_periodo"] = df["semanas_disponibles"] * df["HORAS SEMANALES"]

# (Opcional) Redondear a 2 decimales
df["semanas_disponibles"] = df["semanas_disponibles"].round(2)
df["horas_disponibles_periodo"] = df["horas_disponibles_periodo"].round(2)
```

```python
df[["Fecha inicio","Fecha fin real","Dias disponibles","semanas_disponibles",
     "HORAS SEMANALES","horas_disponibles_periodo"]].head()
```

Con las horas ya calculadas podemos notificar si cumple o no con las horas aprobadas. Primero definimos la condición

```python
# Condición: horas disponibles < horas aprobadas (estrictamente menor)
cond = (
    df["horas_disponibles_periodo"].notna() &
    df["HORAS APROBADAS"].notna() &
    (df["horas_disponibles_periodo"] < df["HORAS APROBADAS"])
)
```

Ahora vamos a establecer la observación a la condición:

```python
msg = "No cumple con las horas aprobadas"
```

Ahora vamos a colocar la observación donde corresponda

```python
# Añadir la observación solo donde aplique y evitando duplicados
pend = cond & ~df["observaciones"].str.contains(re.escape(msg), na=False)
df.loc[pend, "observaciones"] = np.where(
    df.loc[pend, "observaciones"].str.strip().eq(""),
    msg,
    df.loc[pend, "observaciones"] + " ; " + msg
)
```

```python
# (Opcional) ver un resumen
print("Filas marcadas:", pend.sum())
df.loc[pend, ["horas_disponibles_periodo","HORAS APROBADAS","observaciones"]].head()
```
